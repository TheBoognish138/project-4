```python
mapped_df = original_df.copy()
```

```python
# Define the mapping
Gender_mapping = {
        0: "Male",
        1: "Female"
}

# Replace values 
mapped_df['Gender'] = mapped_df['Gender'].map(Gender_mapping)

# Display the updated DataFrame
mapped_df.head()
```

```python
# Define the mapping 
ethnicity_mapping = {
    0: 'Caucasian',
    1: 'African American',
    2: 'Asian',
    3: 'Other'
}

# Replace values 
mapped_df['Ethnicity'] = mapped_df['Ethnicity'].map(ethnicity_mapping)

# Display the updated DataFrame
mapped_df.head()
```

```python
# Define the mapping 
SocioeconomicStatus_mapping = {
        0: 'Low',
        1: 'Middle',
        2: 'High',
}

# Replace values 
mapped_df['SocioeconomicStatus'] = mapped_df['SocioeconomicStatus'].map(SocioeconomicStatus_mapping)

# Display the updated DataFrame
mapped_df.head()
```

```python
# Define the mapping
EducationLevel_mapping = {
        0: "None",
        1: "High School",
        2: "Bachelor's",
        3: "Higher"
}

# Replace values
mapped_df['EducationLevel'] = mapped_df['EducationLevel'].map(EducationLevel_mapping)

# Display the updated DataFrame
mapped_df.head()
```

```python
# Define the mapping
yes_no_mapping = {
        0: "No",
        1: "Yes"
}

# Replace values 
mapped_df['Smoking'] = mapped_df['Smoking'].map(yes_no_mapping)
mapped_df['FamilyHistoryKidneyDisease'] = mapped_df['FamilyHistoryKidneyDisease'].map(yes_no_mapping)
mapped_df['FamilyHistoryHypertension'] = mapped_df['FamilyHistoryHypertension'].map(yes_no_mapping)
mapped_df['FamilyHistoryDiabetes'] = mapped_df['FamilyHistoryDiabetes'].map(yes_no_mapping)
mapped_df['PreviousAcuteKidneyInjury'] = mapped_df['PreviousAcuteKidneyInjury'].map(yes_no_mapping)
mapped_df['UrinaryTractInfections'] = mapped_df['UrinaryTractInfections'].map(yes_no_mapping)
mapped_df['ACEInhibitors'] = mapped_df['ACEInhibitors'].map(yes_no_mapping)
mapped_df['Diuretics'] = mapped_df['Diuretics'].map(yes_no_mapping)
mapped_df['Statins'] = mapped_df['Statins'].map(yes_no_mapping)
mapped_df['AntidiabeticMedications'] = mapped_df['AntidiabeticMedications'].map(yes_no_mapping)
mapped_df['Edema'] = mapped_df['Edema'].map(yes_no_mapping)
mapped_df['HeavyMetalsExposure'] = mapped_df['HeavyMetalsExposure'].map(yes_no_mapping)
mapped_df['OccupationalExposureChemicals'] = mapped_df['OccupationalExposureChemicals'].map(yes_no_mapping)

mapped_df['Diagnosis'] = mapped_df['Diagnosis'].map(yes_no_mapping)

# Display the updated DataFrame
mapped_df.head()
```

```python
# Define the mapping
WaterQuality_mapping = {
        0: "Good",
        1: "Poor"
}

# Replace values
mapped_df['WaterQuality'] = mapped_df['WaterQuality'].map(WaterQuality_mapping)

# Display the updated DataFrame
mapped_df.head()
```

```python
mapped_df.to_csv('Resources/mapped_df.csv', index=False)
```
